```python
def entropy(y):
  values,counts = np.unique(y, return_counts=True)
  probs = counts / counts.sum()
  return -np.sum(probs * np.log2(probs))

def information_gain(df, feature, target):
  total_entropy = entropy(df[target])
  values,counts = np.unique(df[feature], return_counts=True)
  weighted_entropy = 0
  for v,c in zip(values,counts):
    subset = df[df[feature] == v]
    weighted_entropy += (c / counts.sum()) * entropy(subset[target])
  return total_entropy - weighted_entropy


```

Gini INdex (CART)

```python
def gini(y):
  values,counts = np.unique(y, return_counts=True)
  probs = counts / counts.sum()
  return 1 - np.sum(probs ** 2)
def gini_index(df, feature, target):
  values,counts = np.unique(df[feature], return_counts=True)
  gini_total = 0
  for v,c in zip(values,counts):
    subset = df[df[feature] == v]
    gini_total += (c / counts.sum()) * gini(subset[target])
  return gini_total
```

```python
target =df.columns[-1]
features = df.columns[:-1]
print("ENtropy")
for f in features:
  print(f,entropy(target))
print("Information Gain (ID3)")
for f in features:
  print(f,information_gain(df,f,target))
print("\nGini Index")
for f in features:
  print(f,gini_index(df,f,target))
```
