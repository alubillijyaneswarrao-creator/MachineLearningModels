```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv('Diwali_Sales_Data.csv')
print(data.head())
print(data.info())
```

```python
data.isnull().sum()
```

```python
gender_sales = data.groupby("Gender")["Amount"].sum()
sns.barplot(x=gender_sales.index, y=gender_sales.values)
plt.title("Sales by Gender")
plt.show()
```

```python
age_sales = data.groupby("Age Group")["Amount"].sum().sort_values(ascending=False)
sns.barplot(x=age_sales.index, y=age_sales.values)
plt.title("Sales by Age Group")
plt.show()
```

```python
state_sales = data.groupby("State")["Amount"].sum().sort_values(ascending=False)
sns.barplot(x=state_sales.index, y=state_sales.values)
plt.xticks(rotation=45)
plt.title("Sales by State")
plt.show()
```

```python
def plot_sales_by_column(data, column, title):
    sales = data.groupby(column)["Amount"].sum().sort_values(ascending=False)
    ax = sns.barplot(x=sales.index, y=sales.values)
    ax.bar_label(ax.containers[0])
    plt.xticks(rotation=45)
    plt.title(title)
    plt.show()
plot_sales_by_column(data, "State", "Sales by State")
plot_sales_by_column(data, "Product_Category", "Sales by Product Category")
```
